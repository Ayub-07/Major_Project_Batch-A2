```python
import os
from PIL import Image
import os
from PIL import Image

input_folder = "indian_coins_dataset"
output_folder = "indian_coins_dataset_compressed"

# Create the main output folder
os.makedirs(output_folder, exist_ok=True)

for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(root, file)

            # Preserve subfolder structure
            relative_path = os.path.relpath(root, input_folder)
            save_dir = os.path.join(output_folder, relative_path)
            os.makedirs(save_dir, exist_ok=True)

            # Open and compress image
            img = Image.open(img_path)

            # Optional: Resize (e.g., 800x800)
            img.thumbnail((800, 800))  # Maintains aspect ratio

            # Save with compression
            save_path = os.path.join(save_dir, file)
            img.save(save_path, optimize=True, quality=70)  # Quality 1-100
            print(f"Compressed: {img_path} → {save_path}")
```
